```python
def remove_outliers_replace_with_median(df):
    for column in df.select_dtypes(include=[np.number]).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3-Q1
        lower_bound = Q1 - (1.5 * IQR)
        upper_bound = Q3 + (1.5 * IQR)

        median = df[column].median()
        df[column] = df[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
    return df
df1 = remove_outliers_replace_with_median(df)
df1
```

# **Dimensionality Reduction with PCA**

```python
df2 = df1.copy()
df2
```

```python
cat_data = df2['Type']
cat_data
```

```python
# Standardize the original data set

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
variables = ['Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Phenols',
       'Flavanoids', 'Nonflavanoids', 'Proanthocyanins', 'Color', 'Hue',
       'Dilution', 'Proline']
df2[variables] = scaler.fit_transform(df2[variables])
df2
```

```python
# only numerical variables

df3 = df2.iloc[:,1:]
df3
```

```python
# Principal Component Analysis (PCA)

from sklearn.decomposition import PCA
pca = PCA()
pc_data = pca.fit_transform(df3)
pc_data = pd.DataFrame(pc_data)

pc_data
```

```python
# PCA data variance
pc_data.var()
```

```python
pca_data = pc_data.iloc[:,0:3]
pca_data
```

```python
# The amount of variance that each PCA explains
pca.explained_variance_
```

```python
# Scree Plot

plt.plot(range(1, len(pca.explained_variance_) + 1), pca.explained_variance_, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue (Explained Variance)')
plt.show()
```

```python
# Transform the original dataset into the principal components

final_pca = pd.concat([pca_data, cat_data], axis=1)
final_pca.columns = ['pc_1', 'pc_2', 'pc_3', 'Type']
final_pca
```

```python
# visualization for the PCA data

plt.figure(figsize = (8,6))
sns.pairplot(final_pca, hue='Type',palette='rainbow')
plt.show()
```

# **Clustering with PCA Data**

```python
# PCA_ data set
df_pca=final_pca.copy()
```

```python
# drop the type variable

df_pca .drop('Type',axis = 1, inplace = True)
df_pca
```

```python
# Visualize the clustering results using appropriate plots.

from sklearn.cluster import KMeans
cluster = KMeans(n_clusters = 3,random_state = 42)
clust_pca = cluster.fit_predict(df_pca)
clust_pca
```

```python
# add the clust_pca variable

df_pca['clust_pca'] = clust_pca
df_pca
```

```python
from sklearn.metrics import silhouette_score
PCA_data = silhouette_score(df_pca,clust_pca)
print("silhouette score_k_2 :",np.round(PCA_data,3))
```

```python
# Visualize the clustering results using appropriate plots.

plt.figure(figsize = (7,5))
plt.scatter(df_pca.iloc[:,0],df_pca.iloc[:,2], c = df_pca['clust_pca'],cmap='rainbow')
plt.show()
```

```python
#4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.
'''
##Clustering on Original Dataset
Dimensionality:
Better suited for low-dimensional data where the original features are directly meaningful.

Interpretability:
Easier to interpret clusters, as they directly relate to the original features, making the results more understandable.

Correlation Between Features:
May suffer from redundancy due to correlated features, which can distort the clustering process.

Computational Cost:
Higher computational cost, especially for large datasets with many features, as clustering must process all dimensions.

Cluster Quality:
Can capture all dimensions, including those with low variance, which may be important for identifying clusters.



##PCA Before Clustering
Dimensionality:
Effective for high-dimensional data, as it reduces the number of features while retaining most of the variance.

Interpretability:
Harder to interpret clusters, as they are defined in terms of principal components, which are linear combinations of original features.


Correlation Between Features:
Removes multicollinearity by transforming correlated features into uncorrelated principal components.

Computational Cost:
Reduces computational cost by lowering the dimensionality of the data.

Cluster Quality:
May lose clustering-relevant dimensions if those dimensions have low variance and are not preserved in the principal components.
'''
```
